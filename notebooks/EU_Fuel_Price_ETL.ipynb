```python
import requests
import pandas as pd

# Direct URL of the file
url = "https://energy.ec.europa.eu/document/download/906e60ca-8b6a-44e7-8589-652854d2fd3f_en?filename=Weekly_Oil_Bulletin_Prices_History_maticni_4web.xlsx"

# Download
response = requests.get(url)
if response.status_code == 200:
    with open("Weekly_Oil_Bulletin_Prices_History.xlsx", "wb") as f:
        f.write(response.content)
    print("File downloaded successfully!")
else:
    print(f"Error during download: {response.status_code}")
```

```python
df = pd.read_excel("Weekly_Oil_Bulletin_Prices_History.xlsx", sheet_name="Prices with taxes")

display(df.head())
```

```python
# I keep the date column + euro95 and diesel columns for each country
cols_to_keep = ['Consumer prices of petroleum products inclusive of duties and taxes'] + \
               [col for col in df.columns if col.endswith('_euro95') or col.endswith('_diesel')]

df_filtered = df[cols_to_keep]

# Rename the date column
df_filtered = df_filtered.rename(columns={'Consumer prices of petroleum products inclusive of duties and taxes': 'Date'})

display(df_filtered.head())
print(df_filtered.columns.tolist())
```

```python
# Deleted the first 2 lines
df_filtered = df_filtered.iloc[2:].reset_index(drop=True)

display(df_filtered.head())
```

```python
# Removed the columns EUR_price_with_tax_euro95 and EUR_price_with_tax_diesel, as they are only relevant for Eurozone countries
df_filtered = df_filtered.drop(columns=['EUR_price_with_tax_euro95', 'EUR_price_with_tax_diesel'])
```

```python
# Remove columns with exchange_rate in the title
df_filtered = df_filtered[df_filtered.columns.drop(list(df_filtered.filter(regex='exchange_rate')))]
```

```python
# I see there are lines at the bottom with text (to be deleted)
df_filtered.tail(30)
```

```python
# Count total rows before filtering
initial_count = len(df_filtered)

# Remove rows where 'Date' column contains NaN values
df_filtered = df_filtered[df_filtered['Date'].notna()]

# Count remaining rows after filtering
final_count = len(df_filtered)

# Calculate number of removed rows
removed_rows = initial_count - final_count

# Display cleaning results
print(f"Number of removed rows: {removed_rows}")
print(f"Remaining rows: {final_count}")
```

```python
# Convert the 'Date' column to datetime format (date only, no time component)
# errors='coerce' converts invalid parsing to NaT (Not a Time)
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'], errors='coerce')
```

```python
# Droped rows where only 'Date' is NaT
df_filtered = df_filtered.dropna(subset=['Date']).reset_index(drop=True)
```

```python
df_filtered_clean = df_filtered.dropna(subset=['Date']).reset_index(drop=True)
print(f"Remaining lines: {len(df_filtered_clean)}")

# Remake the melt properly
df_sp95 = df_filtered_clean.melt(id_vars='Date',
                                  value_vars=[c for c in df_filtered_clean.columns if c.endswith('_SP95')],
                                  var_name='Country', value_name='SP95')
df_sp95['Country'] = df_sp95['Country'].str.replace('_SP95', '')

df_diesel = df_filtered_clean.melt(id_vars='Date',
                                    value_vars=[c for c in df_filtered_clean.columns if c.endswith('_diesel')],
                                    var_name='Country', value_name='Diesel')
df_diesel['Country'] = df_diesel['Country'].str.replace('_diesel', '')

df_long = df_sp95.merge(df_diesel, on=['Date', 'Country'])
print(df_long.shape)
print(df_long.isna().sum())
```

```python
# I keep only the SP95 columns and reshape them from wide to long format
df_sp95 = df_filtered_clean.melt(
    id_vars='Date',
    value_vars=[c for c in df_filtered_clean.columns if c.endswith('_SP95')],
    var_name='Country',
    value_name='SP95'
)

# I remove '_SP95' from country names so 'France_SP95' becomes 'France'
df_sp95['Country'] = df_sp95['Country'].str.replace('_SP95', '')

# I do the same thing for Diesel columns
df_diesel = df_filtered_clean.melt(
    id_vars='Date',
    value_vars=[c for c in df_filtered_clean.columns if c.endswith('_diesel')],
    var_name='Country',
    value_name='Diesel'
)

# I remove '_diesel' from country names so 'France_diesel' becomes 'France'
df_diesel['Country'] = df_diesel['Country'].str.replace('_diesel', '')

# I merge the two dataframes to get SP95 and Diesel in the same row
df_long = df_sp95.merge(df_diesel, on=['Date', 'Country'])

# I check the result
print(df_long.head(10))
```

```python
# Correction dictionary
corrections = {
    "UnitedKingdom": "United Kingdom",
    "CzechRepublic": "Czech Republic",
    "EuropeanUnion": "European Union"
}

# Applied the corrections
df_long["Country"] = df_long["Country"].replace(corrections)

# Verification
print(df_long["Country"].unique())
```

```python
# Sorted lines by date
df_long = df_long.sort_values('Date', ascending=False).reset_index(drop=True)
df_long.head(10)
```

```python
# Conversion of SP95 and Diesel columns into numbers
df_long['SP95'] = pd.to_numeric(df_long['SP95'], errors='coerce')
df_long['Diesel'] = pd.to_numeric(df_long['Diesel'], errors='coerce')

df_long.info()
```

```python
import os
import json
import uuid
import shutil
import pyarrow as pa
import pyarrow.parquet as pq
from datetime import datetime
import pandas as pd

# 1. Convert the Date column to datetime
df_long['Date'] = pd.to_datetime(df_long['Date'])

# 2. Define the PyArrow schema explicitly with timestamp
schema_arrow = pa.schema([
    pa.field("Date", pa.timestamp('us')),
    pa.field("Country", pa.string()),
    pa.field("SP95", pa.float64()),
    pa.field("Diesel", pa.float64()),
])

# 3. Create the folders (without overwriting)
table_path = "/lakehouse/default/Tables/EU_Fuel_Price_Table"
os.makedirs(table_path, exist_ok=True)
os.makedirs(f"{table_path}/_delta_log", exist_ok=True)

# 4. Write the Parquet with a fixed name
arrow_table = pa.Table.from_pandas(df_long, schema=schema_arrow, preserve_index=False)
parquet_filename = "data.parquet"              
parquet_path = f"{table_path}/{parquet_filename}"
pq.write_table(arrow_table, parquet_path)

# 5. Create the delta_log
def pa_type_to_delta(t):
    if pa.types.is_timestamp(t): return "timestamp"
    if pa.types.is_int64(t) or pa.types.is_int32(t): return "long"
    if pa.types.is_float64(t) or pa.types.is_float32(t): return "double"
    if pa.types.is_date(t): return "date"
    return "string"

schema = arrow_table.schema
delta_schema = {
    "type": "struct",
    "fields": [
        {"name": f.name, "type": pa_type_to_delta(f.type), "nullable": True, "metadata": {}}
        for f in schema
    ]
}

commit = [
    json.dumps({"protocol": {"minReaderVersion": 1, "minWriterVersion": 2}}),
    json.dumps({"metaData": {
        "id": str(uuid.uuid4()),
        "format": {"provider": "parquet", "options": {}},
        "schemaString": json.dumps(delta_schema),
        "partitionColumns": [],
        "createdTime": int(datetime.now().timestamp() * 1000),
        "configuration": {}
    }}),
    json.dumps({"add": {
        "path": parquet_filename,
        "size": os.path.getsize(parquet_path),
        "partitionValues": {},
        "modificationTime": int(datetime.now().timestamp() * 1000),
        "dataChange": True,
        "stats": json.dumps({"numRecords": len(df_long)})
    }})
]

with open(f"{table_path}/_delta_log/00000000000000000000.json", "w") as f:
    f.write("\n".join(commit))

print("✅ Delta table updated without changing the files!")
print(f"Plan: {schema}")
```
